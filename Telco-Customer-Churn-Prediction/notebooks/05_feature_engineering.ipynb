### Binary Encoding

Binary categorical features were converted from text values (Yes/No) into numerical values (1/0).

Encoding was applied to customer demographics, service subscriptions, and account-related features.

This transformation is necessary because machine learning algorithms operate on numerical data and cannot directly interpret categorical text values.

```python
data["churn_label"] = data["churn_label"].map({
    "Yes": 1,
    "No": 0
})

data["churn_label"].value_counts()
```

### Target Encoding

The target variable (churn_label) was converted into a binary numerical format:

- Yes → 1 (Customer Churned)
- No → 0 (Customer Retained)

This transformation is required because machine learning algorithms operate on numerical values rather than text labels.

```python
data["avg_revenue_per_month"] = (
    data["total_revenue"] /
    data["tenure_in_months"]
)

data[
    [
        "total_revenue",
        "tenure_in_months",
        "avg_revenue_per_month"
    ]
].head()
```

### Feature 1: Average Revenue Per Month

This feature was created by dividing Total Revenue by Tenure in Months.

It represents the average monthly revenue generated from each customer over their entire relationship with the company.

Unlike Monthly Charge, which reflects the current billing amount, this feature captures the long-term financial behavior of the customer.

This helps provide a more stable and historical view of customer value.

```python
contract_risk_map = {
    "Month-to-Month": 3,
    "One Year": 2,
    "Two Year": 1
}

data["contract_risk_score"] = data["contract"].map(contract_risk_map)

data[["contract", "contract_risk_score"]].head()
```

```python
data["contract"].value_counts()
```

### Feature 2: Contract Risk Score

The contract type was transformed into a numerical risk score based on churn behavior observed during EDA.

- Month-to-Month → High Risk (3)
- One Year → Medium Risk (2)
- Two Year → Low Risk (1)

This encoding introduces an ordinal relationship between contract types, capturing customer commitment level more effectively than one-hot encoding.

```python
service_cols = [
    "online_security",
    "online_backup",
    "device_protection_plan",
    "premium_tech_support",
    "streaming_tv",
    "streaming_movies",
    "streaming_music",
    "multiple_lines",
    "phone_service",
    "internet_service",
    "unlimited_data"
]

data["service_engagement_score"] = data[service_cols].sum(axis=1)
```

```python
data[["service_engagement_score"] + service_cols].head()
```

### Feature 3: Service Engagement Score

Since binary encoding was already applied (Yes/No → 1/0), the feature was created by directly summing service-related columns.

This feature represents the total number of services used by each customer.

Higher values indicate stronger engagement with the company’s services, while lower values suggest weaker engagement and higher churn risk.

```python
data["referral_engagement_score"] = (
    data["referred_a_friend"] + data["number_of_referrals"]
)
```

```python
data[["referred_a_friend", "number_of_referrals", "referral_engagement_score"]].head()
```

### Feature 4: Referral Engagement Score

This feature combines two variables:
- Whether the customer referred a friend
- The total number of referrals made

The goal is to measure customer loyalty and engagement through word-of-mouth behavior.

Higher values indicate highly satisfied and loyal customers who actively promote the service, while lower values indicate weak or no engagement in referrals.

```python
data["value_stability_score"] = (
    data["avg_revenue_per_month"] - data["monthly_charge"]
)
```

```python
data[[
    "monthly_charge",
    "avg_revenue_per_month",
    "value_stability_score"
]].head()
```

### Feature 5: Value Stability Score

This feature measures the difference between a customer's current monthly charge and their historical average monthly revenue.

It helps capture changes in customer spending behavior over time, which is a key indicator of churn risk.

- Positive values may indicate declining customer value.
- Negative values may indicate growth in customer spending.
- Values close to zero indicate stable customer behavior.

```python
service_groups = {
    "security": ["online_security", "online_backup", "device_protection_plan", "premium_tech_support"],
    "streaming": ["streaming_tv", "streaming_movies", "streaming_music"],
    "connectivity": ["phone_service", "multiple_lines", "internet_service", "unlimited_data"]
}

data["service_breadth_score"] = 0

for group, cols in service_groups.items():
    data[group] = data[cols].sum(axis=1)
    data["service_breadth_score"] += (data[group] > 0).astype(int)
```

```python
data[[
    "security",
    "streaming",
    "connectivity",
    "service_breadth_score"
]].head()
```

### Feature 6: Service Breadth Score

This feature measures how diversified a customer's service usage is across different categories (security, streaming, and connectivity).

Unlike engagement score, which measures quantity, this feature captures variety in usage.

Higher diversity indicates stronger customer integration with the company ecosystem and lower churn risk.

```python
new_features = [
    "avg_revenue_per_month",
    "contract_risk_score",
    "service_engagement_score",
    "referral_engagement_score",
    "value_stability_score",
    "service_breadth_score"
]

data[new_features].describe().round(2)
```

```python
corr_matrix = data.select_dtypes(
    include=["int64", "float64"]
).corr()

corr_matrix["churn_label"].sort_values(
    key=abs,
    ascending=False
)
```

## Feature Engineering Summary

### Retained Features

The following engineered features were retained because they demonstrated meaningful relationships with customer churn and added business value beyond the original variables:

- **Avg Revenue Per Month**
  - Measures the average revenue generated by a customer per month.
  - Helps capture customer value independently of tenure length.

- **Contract Risk Score**
  - Converts contract types into a risk-based scale.
  - Customers with Month-to-Month contracts are assigned higher risk scores than customers with long-term contracts.

- **Referral Engagement Score**
  - Combines referral behavior and the number of referrals made by a customer.
  - Represents customer engagement and loyalty to the company.

---

### Removed Features

The following engineered features were removed because they showed very weak relationships with churn and did not provide sufficient predictive value:

- **Service Engagement Score**
- **Service Breadth Score**
- **Value Stability Score**

Removing these features helps reduce unnecessary complexity while improving model interpretability.

---

### Key Takeaway

Feature engineering should focus on creating variables that capture meaningful customer behavior and business patterns.

After evaluating the engineered features, only the features that demonstrated useful relationships with churn were retained for modeling.

```python
features_to_drop = [
    "service_engagement_score",
    "service_breadth_score",
    "value_stability_score"
]

data = data.drop(columns=features_to_drop)
```

## Categorical Encoding

The remaining categorical features contain multiple categories and cannot be represented using binary encoding.

One-Hot Encoding will be applied to:

- Contract
- Payment Method
- Internet Type
- Offer

The City feature will be removed because it contains a very high number of unique categories (high cardinality), which could introduce unnecessary complexity and increase the dimensionality of the dataset.

```python
categorical_cols = [
    "contract",
    "payment_method",
    "internet_type",
    "offer"
]

data = pd.get_dummies(
    data,
    columns=categorical_cols,
    drop_first=True,
    dtype=int
)
```
